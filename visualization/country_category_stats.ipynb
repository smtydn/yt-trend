```python
import os
data_path = os.path.abspath(
    os.path.join(os.path.abspath(os.path.curdir), os.pardir, "results", "country-category-stats.json"))
print(f"Data Path: {data_path}")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})

%matplotlib widget
```

```python
import json
from pprint import pprint

data = []
with open(data_path) as f:
    for line in f:
        data.append(json.loads(line))
pprint(data[:2])
```

```python
from enum import IntEnum

class Country(IntEnum):
    CA = 0
    DE = 1
    FR = 2
    GB = 3
    IN = 4
    JP = 5
    KR = 6
    MX = 7
    RU = 8
    US = 9
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.head
```

```python
ca = df["country"] == Country.CA
df[ca][["title", "count"]].plot(
    kind="barh", 
    x="title", 
    y="count",
    xlabel="Category",
    ylabel="# of videos",
    title="Numbers of videos for a category that was in trends for CA",
    grid=True,
    fontsize=9
)
```
