```python
import os
data_path = os.path.abspath(
    os.path.join(os.path.abspath(os.path.curdir), os.pardir, "results", "like-ratio.json"))
print(f"Data Path: {data_path}")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})

%matplotlib widget
```

```python
import json
from pprint import pprint

data = []
with open(data_path) as f:
    for line in f:
        data.append(json.loads(line))
pprint(data[:2])
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.head
```

```python
df.plot(
    kind="barh", 
    x="category_title", 
    y="ratio",
    xlabel="Category",
    ylabel="Ratio",
    title="Ratio per category",
    grid=True,
    fontsize=9,
    legend=False,
    figsize=(9, 15)
)
```
